```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = pd.read_csv("datasets/Iklan_sosmed.csv")
datasets.info()
```

```python
X = datasets.iloc[:, [2, 3]].values
y = datasets.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, test = train_test_split(
    X, y, test_size=0.25, random_state=1234
)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=1234)
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
fn = ['Usia', 'EstimasiGaji']
cn = ['0','1']
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(classifier, feature_names=fn, class_names=True);
```

```python
y_pred = classifier.predict(X_test)
ikin = X_test[[12]]
```
